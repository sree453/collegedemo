```python
import numpy as np
import pandas as pd
import difflib  #get the closest match of something
from sklearn.feature_extraction.text import TfidfVectorizer  #converting textual data into numerical values
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection And Data PreProcessing

```python
#loading the data from csv file to pandas dataframe
movies_data=pd.read_csv('/content/movies.csv')
```

```python
#printing first 5 rows of dataframe
movies_data.head()
```

```python
#number of rows and columns in the dataset
movies_data.shape
```

```python
#selecting relevant features for recommendation
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacig the missing values or none values
for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining all the selected features
combined_features=movies_data['genres']+movies_data['keywords']+movies_data['tagline']+movies_data['cast']+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting text data into (numerical data)feature vector.Finding the similarity value for numerical ways is easy than the features by using the cosine similarity
#converting this data into numerical value by using the tfidvectorizer
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
#getting the similarity scores using cosine similarity

similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
# this is providing similarity between all movies in the dataset
```

```python
print(similarity.shape)
```

```python
#getting the movie name
movie_name=input("Enter your favourite movie name: ")
```

```python
#creating list all movies names given in dataset
list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
```

```python
print(close_match)
```

```python
#finding the index of the movie with title
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index
print('Movies suggested for you \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<31):
    print(i,'-',title_from_index)
    i+=1
```

Movie Recommendation System


```python

movie_name=input("Enter your favourite movie name: ")
#creating list all movies names given in dataset
list_of_all_titles=movies_data['title'].tolist()

#finding the close match for movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)

close_match=find_close_match[0]

#finding the index of the movie with title
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]

#getting a list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))


#sorting the movies based on their similarity
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

#print the name of similar movies based on the index
print('Movies suggested for you \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<31):
    print(i,'-',title_from_index)
    i+=1
```
